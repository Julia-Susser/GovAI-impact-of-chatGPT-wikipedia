```python
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime

def get(url):
    headers = {'User-Agent': 'CoolBot/0.0 (https://example.org/coolbot/; coolbot@example.org) generic-library/0.0 Cool Bot (jesgarbage@gmail.com)'}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print("Error fetching data:", response.status_code)
        return None

def fetch_pageviews(start_date, end_date,country):
    url = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{country}.wikipedia.org/mobile-app/user/monthly/{start_date}/{end_date}"
    return get(url)
    
def fetch_edits(start_date, end_date,country):
    url = f"https://wikimedia.org/api/rest_v1/metrics/edits/aggregate/{country}.wikipedia/user/content/monthly/{start_date}/{end_date}"
    print(url)
    return get(url)


def plot(views, edits, country):
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    if views:
        timestamps_views = [item['timestamp'] for item in views['items']]
        views_data = [item['views'] for item in views['items']]
        dates_views = [datetime.strptime(timestamp, '%Y%m%d%H') for timestamp in timestamps_views]

        axs[0].plot(dates_views, views_data)
        axs[0].set_title(f"{country} Page Views")
        axs[0].set_xlabel('Date')
        axs[0].set_ylabel('Page Views')
        axs[0].tick_params(axis='x', rotation=45)
    
    if edits:
        timestamps_edits = [item['timestamp'] for item in edits['items'][0]["results"]]
        edits_data = [item['edits'] for item in edits['items'][0]["results"]]
        dates_edits = [datetime.fromisoformat(timestamp.replace('Z', '+00:00')) for timestamp in timestamps_edits]

        axs[1].plot(dates_edits, edits_data)
        axs[1].set_title(f"{country} Edits")
        axs[1].set_xlabel('Date')
        axs[1].set_ylabel('Edits')
        axs[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()
    
def main():
    # Define the start and end dates for the past year
    start_date = '2016010100'
    end_date = '2022010100'
    for code,country in language_codes:
        # Fetch page views data
        views = fetch_pageviews(start_date, end_date,code)
        edits = fetch_edits(start_date, end_date,code)
        plot(views,edits,country)


if __name__ == "__main__":
    main()
```
