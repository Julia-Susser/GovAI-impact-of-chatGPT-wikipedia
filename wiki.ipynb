```python
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime

def fetch_pageviews(start_date, end_date,country):
    url = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{country}.wikipedia.org/mobile-app/user/daily/{start_date}/{end_date}"
    headers = {'User-Agent': 'CoolBot/0.0 (https://example.org/coolbot/; coolbot@example.org) generic-library/0.0 Cool Bot (jesgarbage@gmail.com)'}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        plot_pageviews(data)
        return data
    else:
        print("Error fetching data:", response.status_code)
        return None
    


def plot_pageviews(data):
    timestamps = [item['timestamp'] for item in data['items']]
    views = [item['views'] for item in data['items']]
    dates = []
    for timestamp in timestamps:
        date_obj = datetime.strptime(timestamp, '%Y%m%d%H')
        dates.append(date_obj)
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=dates, y=views)
    plt.title('Smooth Average Daily Page Views Over the Past Year')
    plt.xlabel('Date')
    plt.ylabel('Page Views')

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
def main():
    # Define the start and end dates for the past year
    start_date = '2010010100'
    end_date = '2022010100'
    for country in ["en"]:
        # Fetch page views data
        data = fetch_pageviews(start_date, end_date,country)


if __name__ == "__main__":
    main()
```
